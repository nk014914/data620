```python
import networkx as net
import matplotlib.pyplot as plot

# Step 1: Create the Krackhardt Kite Graph
G = net.Graph()

# Add nodes
nodes = ["Alex", "Lee", "Sophia", "Elliot", "Elise", "Skyler"]
G.add_nodes_from(nodes)

# Add edges (connections between nodes)
edges = [
    ("Alex", "Lee"),
    ("Alex", "Sophia"),
    ("Lee", "Sophia"),
    ("Lee", "Elliot"),
    ("Sophia", "Elise"),
    ("Elliot", "Skyler"),
    ("Elise", "Skyler"),
]
G.add_edges_from(edges)

# Step 2: Determine node colors based on degree
degree_dict = dict(G.degree())
max_degree = max(degree_dict.values())
node_colors = ['orange' if degree_dict[node] == max_degree else 'skyblue' for node in G.nodes()]

# Step 3: Draw the graph
plot.figure(figsize=(8,6))
pos = net.spring_layout(G)  # nicer layout
net.draw(
    G,
    pos,
    with_labels=True,
    node_color=node_colors,
    node_size=1500,
    font_size=12,
    font_weight='bold',
    edge_color='gray'
)
plot.title("Hello, Graph World")
plot.show()
```
